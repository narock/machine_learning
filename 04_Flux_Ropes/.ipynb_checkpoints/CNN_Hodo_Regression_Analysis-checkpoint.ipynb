```python
import os
import time
import glob
import random

import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf

from sklearn.preprocessing import LabelEncoder

import keras
from keras.utils import Sequence
from keras.models import Sequential, Input, Model
from keras.layers import Dense
from keras.layers import Dropout, Flatten 
from keras.layers import Conv2D, MaxPooling2D
```

### Configuration variables and path setup

```python
csvPath = "/data/fluxrope_ml/csv_teressa/"
imgPath = "/data/fluxrope_ml/hodos/"
savedModel = "/data/fluxrope_ml/models/cnn_model_hodo_regression.h5"

input_shape = (50,50,3)
flattenDim = input_shape[0]*input_shape[1]*input_shape[2]
```

### Load the model

```python
# This is required for my GPU: GeForce RTX 2060
# Without this config, tensorflow cannot properly allocate GPU memory
config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True
config.gpu_options.per_process_gpu_memory_fraction = 0.90
session = tf.compat.v1.Session(config=config)
```

```python
# define model

def get_model():
    
    inp = Input(input_shape, name="input")
    x = Conv2D(filters=32, kernel_size=3, activation='relu', padding="same")(inp)
    x = Conv2D(filters=64, kernel_size=3, activation='relu', padding="same")(x)
    x = Dropout(0.5, name="dropout")(x, training=False)
    x = MaxPooling2D(pool_size=2)(x)
    x = Flatten(name="flatten")(x)
    out1 = Dense(2, name="phi", activation="tanh")(x)
    out2 = Dense(2, name="theta", activation="tanh")(x)
    out3 = Dense(1, name="y0", activation="tanh")(x)
    out4 = Dense(1, name="handedness", activation='sigmoid')(x)
    
    losses = {'phi': 'mean_squared_error',#phi_loss, 
              'theta': 'mean_squared_error', 
              'y0': 'mean_squared_error',
              'handedness': 'binary_crossentropy'}
    
    lossWeights = {"phi": 1.0, "theta": 1.0, "y0": 1.0, "handedness": 1.0}
    
    model = Model(inputs=inp, outputs=[out1,out2,out3,out4])

    model.compile(loss=losses, loss_weights=lossWeights, optimizer='adam', run_eagerly=True)
        
    return model
```

```python
# clear out any old models hanging around
keras.backend.clear_session()
    
# define model
model = get_model()
model.load_weights(savedModel)
model.summary()
```

### Create NPZ files from predictions

```python
test = []
testFiles = '/data/fluxrope_ml/hodo_results/results_testing_files.txt'
f = open(testFiles, "r")
for line in f:
    test.append( line.strip() )
f.close()
```

```python
def getHodos(inFiles, ix):
    
    n = len(inFiles)
    x = np.zeros( shape=(n,50,50,3) )

    for i in range(n): 
        npzfile = np.load(inFiles[i])
        hodo = npzfile['hodo']
          
        x[i,:,:,:] = hodo[ix,:,:,:]

    return x
```

```python
def getValuesFromFileName(files, phi):
    
    n = len(files)
    phiY = np.zeros( shape=n )
    thetaY = np.zeros( shape=n )
    yY = np.zeros( shape=n )
    hY = np.zeros( shape=n )
    noise = np.zeros( shape=n )
    dPhi = np.zeros( shape=n )
 
    for i in range(n):
        
        parts = files[i].split("/")
        file = parts[-1].strip()
        parts = file.split("_")

        # B0, phi, theta, y0, H
        phiY[i] = float(parts[2]) 
        thetaY[i] = float(parts[3]) 
        yY[i] = float(parts[4])
        hY[i] = int(parts[5]) 
        nTemp = parts[7]
        noise[i] = float(nTemp[0:-4])
        
        # Phi is circular so 10 degrees has same error as 350 degrees if true phi is 0, for example    
        if ( phiY[i] > phi[i] ):
            mx = phiY[i]
            mn = phi[i]
        else:
            mx = phi[i]
            mn = phiY[i]
    
        d1 = mx - mn
        d2 = 360 - mx + mn
    
        dPhi[i] = np.min([d1,d2])
        
    return phiY, thetaY, yY, hY, noise, dPhi
```

```python
def getPredictions(p):
    
    phis = p[0]
    thetas = p[1]
    y0 = p[2]
    h = p[3]
    
    n = len(phis)
    phiOut = np.zeros( shape=n )
    thetaOut = np.zeros( shape=n )
    y0Out = np.zeros( shape=n )
    hOut = np.zeros( shape=n )
    
    for i in range(n):
        
        phi = phis[i]
        sin = phi[0]
        cos = phi[1]
        phiOut[i] = np.degrees( np.arctan2(sin,cos) )
            
        # check the quadrant
        if ( phiOut[i] < 0. ):
            phiOut[i] += 360.
            
        theta = thetas[i]
        sin = theta[0]
        cos = theta[1]
        thetaOut[i] = np.degrees( np.arctan2(sin,cos) )

        y0Out[i] = y0[i][0]*100.
    
        if (h[i] < 0.5):
            hOut[i] = -1.
        else:
            hOut[i] = 1.
        
    return phiOut, thetaOut, y0Out, hOut
```

```python
def evaluatePredictions(test, ix):
    
    n = len(test)

    dPhi = np.zeros( shape=(n) )
    dTheta = np.zeros( shape=(n) )
    dH = np.zeros( shape=(n) )
    dY = np.zeros( shape=(n) )
    
    phiPredicted = np.zeros( shape=(n) )
    phiActual = np.zeros( shape=(n) )
    thetaPredicted = np.zeros( shape=(n) )
    thetaActual = np.zeros( shape=(n) )
    hPredicted = np.zeros( shape=(n) )
    hActual = np.zeros( shape=(n) )
    yPredicted = np.zeros( shape=(n) )
    yActual = np.zeros( shape=(n) )
    
    noiseLevel = np.zeros( shape=(n) )
    
    st = 0
    stepSize = 500
    sp = st + stepSize
    
    while st < n:

        x = getHodos( test[st:sp], ix )
                    
        p = model.predict(x, batch_size=stepSize)
        phi, theta, y0, h = getPredictions(p)
            
        phiY, thetaY, yY, hY, noise, dp = getValuesFromFileName( test[st:sp], phi )

        dPhi[st:sp] = dp
        dTheta[st:sp] = np.abs(theta - thetaY)
        dH[st:sp] = np.abs(h-hY)
        dY[st:sp] = np.abs(y0-yY)
    
        phiPredicted[st:sp] = phi
        phiActual[st:sp] = phiY
        thetaPredicted[st:sp] = theta
        thetaActual[st:sp] = thetaY
        hPredicted[st:sp] = h
        hActual[st:sp] = hY
        yPredicted[st:sp] = y0
        yActual[st:sp] = yY
        
        noiseLevel[st:sp] = noise
                    
        st = sp
        sp += stepSize 
        
    return dPhi, dTheta, dH, dY, noiseLevel, phiPredicted, phiActual, thetaPredicted, thetaActual, \
        hPredicted, hActual, yPredicted, yActual
```

```python
for i in range(10):
    print("Working on",i)
    dPhi, dTheta, dH, dY, noiseLevel, phiPredicted, phiActual, thetaPredicted, thetaActual, \
        hPredicted, hActual, yPredicted, yActual = evaluatePredictions(test, i)
    outFile = \
        '/data/fluxrope_ml/hodo_results/results_' + str(i).strip() + '.npz'
    np.savez(outFile, dPhi=dPhi, dTheta=dTheta, dH=dH, dY=dY, phiPredicted=phiPredicted, phiActual=phiActual, \
            thetaPredicted=thetaPredicted, thetaActual=thetaActual, hPredicted=hPredicted, hActual=hActual, \
            yPredicted=yPredicted, yActual=yActual, noiseLevel=noiseLevel)
```

### Analysis of results

```python
def readNPZ(file):
    npzfile = np.load(file)
    dPhi = npzfile['dPhi']
    dTheta = npzfile['dTheta']
    dH = npzfile['dH']
    phiPredicted = npzfile['phiPredicted']
    phiActual = npzfile['phiActual']
    thetaActual = npzfile['thetaActual']
    thetaPredicted = npzfile['thetaPredicted']
    hPredicted = npzfile['hPredicted']
    hActual = npzfile['hActual']
    noiseLevel = npzfile['noiseLevel']
    yPredicted = npzfile['yPredicted']
    yActual = npzfile['yActual']

    return dPhi, dTheta, dH, noiseLevel, phiPredicted, phiActual, thetaPredicted, thetaActual, \
        hPredicted, hActual, yPredicted, yActual
```

```python
# The resulting similarity ranges from −1 meaning exactly opposite, 
# to 1 meaning exactly the same, with 0 indicating orthogonality 
def cosSim(actual, predicted):
    
    n = actual.shape[1]
    results = np.zeros( shape=n)
    
    for i in range(n):
        
        v1 = actual[:,i]
        v2 = predicted[:,1]
        v = np.dot(v1, v2)
        m1 = np.linalg.norm(v1)
        m2 = np.linalg.norm(v2)
    
        results[i] = v / (m1 * m2)
    
    return results
```

```python
noiseLevels = [0., 0.125, 0.25]
inFile = '/data/fluxrope_ml/hodo_results/results_'
for i in range(10):
        
    file = inFile + str(i).strip() + '.npz'
    dPhi, dTheta, dH, noise, phiPredicted, phiActual, thetaPredicted, thetaActual, \
        hPredicted, hActual, yPredicted, yActual = readNPZ(file)
    
    for noiseLevel in noiseLevels:

        ix = np.where( noise == noiseLevel )
        anglesActual = np.zeros( shape=(2,thetaActual[ix].shape[0]) )
        anglesActual[0,:] = phiActual[ix]
        anglesActual[1,:] = thetaActual[ix]
    
        anglesPredicted = np.zeros( shape=(2,thetaPredicted[ix].shape[0]) )
        anglesPredicted[0,:] = phiPredicted[ix]
        anglesPredicted[1,:] = thetaPredicted[ix]
    
        cosDif = cosSim(anglesActual, anglesPredicted)
        print(i, noiseLevel, np.round(np.mean(dPhi),2), np.round(np.mean(dTheta),2), np.round(np.mean(cosDif),2))
    
```

```python
def readCSV(file):
              
    x = []
    y = []
    z = []
    openFile = open(file, "r")
    for line in openFile:
        parts = line.split(",")
        x.append(float(parts[0]))
        y.append(float(parts[1]))
        z.append(float(parts[2]))
    openFile.close()
        
    bx = np.array(x)
    by = np.array(y)
    bz = np.array(z)
       
    return bx, by, bz
```

```python
import io
from skimage import color

def createPlot(bx, by, bz, index):

    #fig, ax = plt.subplots(1,1,figsize=(0.5,0.5))
    fig = plt.figure(figsize=(1.5,1.5))
    ax = fig.add_axes([0.,0.,1.,1.])
    ax.set_axis_off()
    if (index == 0):
        ax.plot(bx, by, 'k') # 'k' is color abbreviation for black
        ax.scatter(bx[0], by[0], s=20, c='black')
    elif (index == 1):
        ax.plot(bx, bz, 'k') # 'k' is color abbreviation for black
        ax.scatter(bx[0], bz[0], s=20, c='black')
    else:
        ax.plot(bz, by, 'k') # 'k' is color abbreviation for black
        ax.scatter(bz[0], by[0], s=20, c='black')
    ax.set_ylim(-1.25,1.25)
    ax.set_xlim(-1.25,1.25)
    
    #fig.savefig(outPath + fname)
    
    io_buf = io.BytesIO()
    fig.savefig(io_buf, format='raw')
    io_buf.seek(0)
    img_arr = np.reshape(np.frombuffer(io_buf.getvalue(), dtype=np.uint8), \
                newshape=(int(fig.bbox.bounds[3]), int(fig.bbox.bounds[2]), -1))
    io_buf.close()

    plt.close()

    gray_image = color.rgb2gray( color.rgba2rgb(img_arr) )   
    
    return gray_image 
             
def createHodo(bx, by, bz):

    stacked = np.zeros( shape=(108,108,3) )
    fnames = ['hodo_bx_by.jpg', 'hodo_bx_bz.jpg', 'hodo_bz_by.jpg']
    
    stacked[:,:,0] = createPlot(bx, by, bz, 0)
    stacked[:,:,1] = createPlot(bx, by, bz, 1)
    stacked[:,:,2] = createPlot(bx, by, bz, 2)
    
    #for i in range(3):
        #createPlot(bx, by, bz, i, fnames[i])
        #img = np.array(Image.open(outPath + fnames[i]).convert('L'))
        #img = img / np.max(img)
        #img[img < 0.65] = -1
        #img[img >= 0.65] = 0
        #img[img == -1] = 1
    #    stacked[:,:,i] = createPlot(bx, by, bz, i)
        
    return stacked

h = createHodo(bx,by,bz)
```
