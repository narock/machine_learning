#### FFT is not machine learning. However, some CNN applications convert time series data to a spectrogram 'image' and then feed that into the neural network. This notebook reviews FFT and spectrograms.

The file piano.txt contains the amplitude data for a note played on a piano. The wave oscillations die off over time as the piano string stops vibrating.

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.loadtxt("piano.txt",float)
x = np.arange(0, len(y), 1)
plt.plot(x/44000,y) # sampling rate is 44,000 times per second
plt.ylabel('Amplitude')
plt.xlabel('Time (sec.)')
plt.title('Piano Note')
```

Apply FFT to convert from amplitude space to frequency space.

```python
import scipy.fftpack
fft = scipy.fftpack.fft(y) 
freqs = scipy.fftpack.fftfreq(len(fft)) * 44000
```

```python
plt.plot( freqs, abs(fft) )
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0,3000)
```

Zooming in on the dominant frequency - the others are harmonics.

```python
plt.plot( freqs, abs(fft) )
plt.xlim(500,550)
```

```python
x = abs(fft)
mx = np.max(x)
x = x.tolist()
i = x.index(mx)
print("The dominant frequency is", freqs[i])
```

According to [this Wikipedia page](https://en.wikipedia.org/wiki/Piano_key_frequencies) this note would be the 52nd key on the piano, or C<sub>5</sub> Tenor C

The FFT tells us the dominant frequencies. However, it doesn't tell us when those frequencies occured. A spectrogram is a three dimensional plot showing frequency intensity as a function of time.

```python
import scipy.signal

freqs, times, Sx = scipy.signal.spectrogram(y, fs=44000)

plt.pcolormesh(times, freqs, 10 * np.log10(Sx), cmap='viridis') # convert to decibel
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.ylim(0, 5000)
```

Spectrogram has many optional parameters that control the windowing and window size. A Hanning window, for example, makes the dominant frequency and the harmonics easier to see in the spectrogram.

```python
freqs, times, Sx = scipy.signal.spectrogram(y, fs=44000, window='hanning',
                                            nperseg=1024, noverlap=924, detrend=False,
                                            scaling='spectrum')

plt.pcolormesh(times, freqs, 10 * np.log10(Sx), cmap='viridis') # convert to decibel
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.ylim(0, 3000)
```

Another example - sunspot data. FFT doesn't have to be applied to a 'real' wave. Any data that is wave-like will work. The number of sunspots increases and decreases and looks like a wave when plotted. Applying FFT can get us the period of the sunspot cycle.

```python
import pandas as pd
url = "http://narock.github.io/teaching/CST-411/sunspots.txt"
data = pd.read_csv(url, sep = '\s+')
data.head()
```

```python
plt.plot(data['index'],data['sunspots'])
plt.xlabel('Month')
plt.ylabel('# Sunspots')
```

```python
fft = scipy.fftpack.fft(data['sunspots'])
freqs = scipy.fftpack.fftfreq(len(fft)) 
plt.stem( 1./freqs, abs(fft) ) # 1 / freq gives us the period
plt.xlim(0,200)
plt.ylabel('Magnitude')
plt.xlabel('Period (Months)')
```

```python
x = abs(fft)
mx = np.max(x[1:]) # skip the first point, artificial peak at fft[0]
x = x.tolist()
i = x.index(mx)
p = 1./freqs[i]
print("The dominant period is at", np.round(p,2), "months")
print("This corresponds to a sunspot cycle of", np.round(p/12.,2), "years")
```

The [Dalton Minimum](https://en.wikipedia.org/wiki/Dalton_Minimum) was a period of low sunspot count, representing low solar activity, named after the English meteorologist John Dalton. The Dalton Minimum lasted from about 1790 to 1830.

Our sunspot data begins in 1749. If we plot the spectrogram, with an x-axis of 'years since 1749', then we should see a decrease in the signal around x-value 41.

```python
freqs, times, Sx = scipy.signal.spectrogram(y, nperseg=256)

plt.subplot(2,1,1)
plt.plot(data['index']/12, data['sunspots'])
plt.xlabel('Years after 1749')
plt.ylabel('# of Sunspots Observed')
plt.title('Solar Sunspot Cycle')
plt.plot([41,41],[0,250],'r-')
plt.plot([81,81],[0,250],'r-')
plt.text(43, 225, "Dalton Minimum")

plt.subplot(2,1,2)
plt.pcolormesh(times/12, freqs, Sx, cmap='viridis') 
plt.ylabel('Frequency (Hz)')
plt.xlabel('Years after 1749')
plt.ylim(0,0.03)
plt.xlim(0, 250)
plt.plot([41,41],[0,0.03],'r-')
plt.plot([81,81],[0,0.03],'r-')

plt.subplots_adjust(hspace = 0.3, left = 0, right = 1.5, bottom = 0, top = 1.5)
```
